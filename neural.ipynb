```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data as TensorDataset
import torch.utils.data as DataLoader
import pandas as pd
import numpy as np
```

```python
y = torch.tensor(0)
one_hot_target = F.one_hot(y, num_classes=3)

scores = torch.tensor([0.005, 0.005, 0.99])

criterion = nn.CrossEntropyLoss()
loss = criterion(scores.double(), one_hot_target.double())

print(loss)
```

```python
input_tensor = torch.tensor([[1.0, 2.0, 3.0]])
y = torch.tensor([0])

model = nn.Sequential(nn.Linear(3, 8), nn.Linear(8, 4), nn.Linear(4, 3))
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()

    predictions = model(input_tensor)
    loss = loss_fn(predictions, y)

    loss.backward()
    optimizer.step()

    result = torch.softmax(predictions, dim=-1)
    print(f"Epoch {epoch}: Loss = {(loss.item()*100):.2f}%, Predictions = {result}")

    if loss.item() < 0.01:
        print("Loss is low, stopping training.")
        break
```

```python
df = pd.read_csv('animals.csv')

X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values

dataset = TensorDataset.TensorDataset(
    torch.tensor(X, dtype=torch.float32),
    torch.tensor(y, dtype=torch.long)
)

input_sample, label_sample = dataset[0]

print(f"Input Sample: {input_sample}")
print(f"Label Sample: {label_sample}")

dataloader = torch.utils.data.DataLoader(dataset, batch_size=2, shuffle=True)
for batch_inputs, batch_labels in dataloader:
    print(f"Batch Inputs:\n{batch_inputs}")
    print(f"Batch Labels:\n{batch_labels}")
    break
```
