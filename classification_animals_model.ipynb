```python
import torch
import torch.nn as nn
import torch.utils.data as TensorDataset
import torch.utils.data as DataLoader
import pandas as pd
```

```python
df = pd.read_csv('animals.csv')

X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values

dataset = TensorDataset.TensorDataset(
    torch.tensor(X, dtype=torch.float32),
    torch.tensor(y, dtype=torch.long)
)

dataloader = DataLoader.DataLoader(dataset, batch_size=20, shuffle=True)

model_animals = nn.Sequential(
    nn.Linear(7, 64),
    nn.Linear(64, 32),
    nn.Linear(32, 3),
)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model_animals.parameters(), lr=0.01)

num_epochs = 100000

for epoch in range(num_epochs):
    for features, labels in dataloader:
        optimizer.zero_grad()
        predictions = model_animals(features)
        loss = criterion(predictions, labels)
        loss.backward()
        optimizer.step()

    if loss.item() < 0.01:
        print("Ошибка минимальна, прекращаем обучение.")
        break
    print(f"Эпоха {epoch}: Ошибка = {(loss.item()*100):.2f}%")
```
